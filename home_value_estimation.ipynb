```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn import metrics
from sklearn.metrics import mean_absolute_error as mae
from sklearn.svm import SVC
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/Zillow.csv')
df.head()
```

```python
df.shape
```

```python
to_remove = []
for col in df.columns:

    if df[col].nunique() == 1:
        to_remove.append(col)
    elif (df[col].isnull()).mean() > 0.60:
        to_remove.append(col)

print(len(to_remove))

```

```python
df.drop(columns=to_remove, inplace=True)
df.info()
```

```python
df.isnull().sum().plot.bar()
plt.show()
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0])
    elif df[col].dtype == np.number:
        df[col] = df[col].fillna(df[col].mean())

df.isnull().sum().sum()
```

```python
ints, objects, floats = [], [], []

for col in df.columns:
    if df[col].dtype == float:
        floats.append(col)
    elif df[col].dtype == int:
        ints.append(col)
    else:
        objects.append(col)

len(ints), len(floats), len(objects)
```

```python
for col in objects:
    print(col, ' -> ', df[col].nunique())
    print(df[col].unique())
    print()
```

```python
plt.figure(figsize=(8, 5))
sb.distplot(df['target'])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sb.boxplot(df['target'])
plt.show()
```

```python
print('Shape of the dataframe before removal of outliers', df.shape)
df = df[(df['target'] > -1) & (df['target'] < 1)]
print('Shape of the dataframe after removal of outliers ', df.shape)
```

```python
for col in objects:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
plt.figure(figsize=(15, 15))
sb.heatmap(df.corr() > 0.8,
           annot=True,
           cbar=False)
plt.show()
```

```python
to_remove = ['calculatedbathnbr', 'fullbathcnt', 'fips',
             'rawcensustractandblock', 'taxvaluedollarcnt',
             'finishedsquarefeet12', 'landtaxvaluedollarcnt']

df.drop(to_remove, axis=1, inplace=True)
```

```python
features = df.drop(['parcelid'], axis=1)
target = df['target'].values

X_train, X_val,\
    Y_train, Y_val = train_test_split(features, target,
                                      test_size=0.1,
                                      random_state=22)
X_train.shape, X_val.shape
```

```python
# Normalizing the features for stable and fast training.
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

```python
from sklearn.metrics import mean_absolute_error as mae
models = [LinearRegression(), XGBRegressor(),
          Lasso(), RandomForestRegressor(), Ridge()]

for i in range(5):
    models[i].fit(X_train, Y_train)

    print(f'{models[i]} : ')

    train_preds = models[i].predict(X_train)
    print('Training Error : ', mae(Y_train, train_preds))

    val_preds = models[i].predict(X_val)
    print('Validation Error : ', mae(Y_val, val_preds))
    print()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor

# Define models
models = [LinearRegression(), XGBRegressor(), Lasso(), RandomForestRegressor(), Ridge()]
model_names = ["Linear Regression", "XGBoost", "Lasso", "Random Forest", "Ridge"]

# Store metrics
train_r2, val_r2 = [], []
train_rmse, val_rmse = [], []
train_mape, val_mape = [], []

# Train models and compute metrics
for i in range(len(models)):
    models[i].fit(X_train, Y_train)

    print(f'{models[i]} : ')

    # Predictions
    train_preds = models[i].predict(X_train)
    val_preds = models[i].predict(X_val)

    # R² Score
    train_r2.append(r2_score(Y_train, train_preds))
    val_r2.append(r2_score(Y_val, val_preds))
    print('Training R² Score : ', train_r2[-1])
    print('Validation R² Score : ', val_r2[-1])

    # RMSE (Root Mean Squared Error)
    train_rmse.append(np.sqrt(mean_squared_error(Y_train, train_preds)))
    val_rmse.append(np.sqrt(mean_squared_error(Y_val, val_preds)))
    print('Training RMSE : ', train_rmse[-1])
    print('Validation RMSE : ', val_rmse[-1])

    # MAPE (Mean Absolute Percentage Error)
    train_mape.append(np.mean(np.abs((Y_train - train_preds) / Y_train)) * 100)
    val_mape.append(np.mean(np.abs((Y_val - val_preds) / Y_val)) * 100)
    print('Training MAPE : ', train_mape[-1], '%')
    print('Validation MAPE : ', val_mape[-1], '%')

    print()

# Function to plot comparative bar charts
def plot_comparison(train_values, val_values, metric_name, ylabel):
    x = np.arange(len(model_names))  # Label locations
    width = 0.35  # Bar width

    fig, ax = plt.subplots(figsize=(8, 5))
    ax.bar(x - width/2, train_values, width, label="Training", color="blue")
    ax.bar(x + width/2, val_values, width, label="Validation", color="orange")

    # Labels, title, and custom x-axis tick labels
    ax.set_xlabel("Models")
    ax.set_ylabel(ylabel)
    ax.set_title(f"Training vs Validation {metric_name} for ML Models")
    ax.set_xticks(x)
    ax.set_xticklabels(model_names, rotation=30)
    ax.legend()

    # Show plot
    plt.show()

# Plot R² Score comparison
plot_comparison(train_r2, val_r2, "R² Score", "R² Score")

# Plot RMSE comparison
plot_comparison(train_rmse, val_rmse, "RMSE", "Root Mean Squared Error")

# Plot MAPE comparison
plot_comparison(train_mape, val_mape, "MAPE", "Mean Absolute Percentage Error (%)")

```
